```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from quotes.models import ChadhNovel, Alignment

def plot_chadh_histogram(novel):
    
    alignments = (
        Alignment.query
        .filter(
            Alignment.a_id==novel.id,
            Alignment.size >= 5
        )
    )
    
    quoted = []
    for a in alignments:
        quoted += range(a.a_start, a.a_start + a.size)

    
    text = Text(novel.text)
    
    plt.figure(figsize=(20, 5))

    n, bins, patches = plt.hist(
        quoted,
        range=(0, len(text.tokens)),
        bins=500,
        color='green',
    )

    print(novel.slug)
    plt.show()
```

```python
from quotes.text import Text

for novel in ChadhNovel.query.all():
    plot_chadh_histogram(novel)
```
